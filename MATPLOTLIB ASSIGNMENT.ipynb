```python
"""Matplotlib is a popular Python library for data visualization.
   It provides a wide range of tools and functions for creating 
   various types of plots, charts, and graphs."""
'''Here are five plots that can be plotted using the Pyplot module of Matplotlib:

Line Plot: A line plot is used to display data points connected by straight lines.
It is commonly used to show the trend or relationship between two variables over a continuous interval.

Bar Plot: A bar plot, also known as a bar chart, is used to compare categorical data
using rectangular bars. Each bar represents a category, and the height of the bar 
indicates the value associated with that category.

Scatter Plot: A scatter plot is used to visualize the relationship between 
two continuous variables. It displays data points as individual dots on a 
graph, where the position of each dot represents the values of the two variables.

Histogram: A histogram is used to visualize the distribution of a continuous
variable. It divides the range of values into bins and displays the frequency
or count of data points falling into each bin.

Pie Chart: A pie chart is used to represent categorical data as a circular graph.
Each category is represented by a slice of the pie, and the size of each slice 
corresponds to the proportion or percentage of that category.'''
```

```python
'''A scatter plot is a type of data visualization that displays the relationship
between two continuous variables. It is particularly useful for identifying patterns
, trends, and correlations between variables. In a scatter plot, each data point is
represented as a single dot or marker on a graph, with one variable plotted along 
the x-axis and the other variable plotted along the y-axis.
'''
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y, c = 'r')
plt.xlabel("values of x")
plt.ylabel("values of y")
plt.title("scatter plot")
plt.show()
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

fig, ax = plt.subplots()
ax.boxplot(data)

ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Values')

plt.title('Box Plot')
plt.show()

```
